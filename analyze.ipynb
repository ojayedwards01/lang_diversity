```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Checking AFROLID predictions for num of correct predictions

yoruba_pred_df = pd.read_csv('afrolid_results_yoruba.csv')
kinyarwanda_pred_df = pd.read_csv('afrolid_results_kinyarwanda.csv')
amharic_pred_df = pd.read_csv('afrolid_results_amharic.csv')

print("Total number of predictions for Yoruba:", len(yoruba_pred_df)    )
print("Total number of predictions for Kinyarwanda:", len(kinyarwanda_pred_df))
print("Total number of predictions for Amharic:", len(amharic_pred_df))
```

```python
yor_pred_count = yoruba_pred_df['predicted_language'].value_counts()

# Plot pie chart with percentages
plt.figure(figsize=(6, 6))
plt.pie(yor_pred_count.values, 
         labels=yor_pred_count.index, 
         autopct='%1.1f%%',
         startangle=90)
plt.title('AfroLID Yoruba Predictions Count (%)')
plt.show()
```

```python
# Afrolid AMHARIC predictions
amharic_pred_count = amharic_pred_df['predicted_language'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(amharic_pred_count.values, 
        labels=amharic_pred_count.index, 
        autopct='%1.1f%%',
        startangle=90)

plt.title('AfroLID Amharic Predictions Count (%)')
plt.show()
```

```python
# Kinyarwanda predictions
kinyarwanda_pred_count = kinyarwanda_pred_df['predicted_language'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(kinyarwanda_pred_count.values, 
        labels=kinyarwanda_pred_count.index, 
        autopct='%1.1f%%',
        startangle=90)

plt.title('AfroLID Kinyarwanda Predictions Count (%)')
plt.show()
```
